<h3 style="text-align: justify;">
Este script apresenta a metodologia utilizada para calcular a taxa de partos cesarianos realizados por beneficiárias de planos privados de saúde. Os dados utilizados foram obtidos no Portal Brasileiro de Dados Abertos a partir das publicações da Agência Nacional de Saúde Suplementar - ANS.

<h3 style="text-align: justify;">
<b><span style="color: #D6BF16;">Metodologia</span></b>
</h3>

<p style="text-align: justify;">Nesta tarefa prática, estamos interessados em obter dados de quantidade de partos realizados em 2024 nos estados brasileiros, para responder os seguintes questionamentos:
</p>

1) Qual a proporção de partos cesáreos por UF?
2) Qual a quantidade e a proporção de partos cesáreos realizados por mês?
3) Qual a proporção de partos cesáreos realizados por faixa etária?
4) Qual a proporção de partos cesáreos por UF, realizados em beneficiárias jovens, com idades entre 15 e 19 anos?
5) Qual a proporção de partos cesáreos por UF, realizados em beneficiárias com idades entre 40 a 49 anos?
6) Qual o valor médio das despesas com partos por UF?


<p style="text-align: justify;">

<b>Indicador de Partos Cesáreos</b>

O objetivo deste exercício é comparar a taxa de partos cesáreos realizados por beneficiárias de planos privados de saúde em 2024. Essa métrica é obtida dividindo-se o total de partos cesarianos pelo total de partos registrados dentro de cada unidade da federação, da seguinte forma:

<b>Fórmula de Cálculo</b>

$$\text{Taxa de Cesarianas} = \frac{\text{Total de Partos Cesáreos}}{\text{Total de Partos (Vaginal + Múltiplo + Cesáreo)}} \times 100$$

</p>

<p style="text-align: justify;">
<b>Filtros</b>

Selecionam-se exclusivamente os eventos ocorridos no ano de 2024, vinculados a planos de segmentação hospitalar com obstetrícia e beneficiárias com idades entre 15 e 49 anos. A identificação dos partos é realizada a partir dos códigos TUSS correspondentes aos procedimentos de parto vaginal, parto múltiplo, cesariana e cesariana com histerectomia.

* Unidade da Federação - UF
* Ano = 2024
* Planos com Segmentação Hospitalar com Obstetrícia
* Faixa etária de 15 a 49 anos
</p>

### Hospitalar Detalhada

```python
%%time

##########################################################################
# BAIXANDO DADOS DE PROCEDIMENTOS HOSPITALARES POR UF - BASES DETALHADAS
##########################################################################

####################################################################################################
# Configuração

# Colunas desejadas
colunas = [
    'ID_EVENTO_ATENCAO_SAUDE',
    'ANO_MES_EVENTO',
    'UF_PRESTADOR',
    'CD_PROCEDIMENTO',
    'CD_TABELA_REFERENCIA',
    'QT_ITEM_EVENTO_INFORMADO',
    'VL_ITEM_EVENTO_INFORMADO',
    'IND_PACOTE'
]

####################################################################################################
# Extração

# Lista para armazenar todos os DataFrames
dfs = []

# Loop pelas UFs e meses
for uf in ufs:
    print(f"\n########## Iniciando UF: {uf} ##########")
    for mes in range(1, 13):
        zip_url = f"https://dadosabertos.ans.gov.br/FTP/PDA/TISS/HOSPITALAR/2024/{uf}/{uf}_2024{mes:02d}_HOSP_DET.zip"
        print(f"Baixando: {zip_url}")

        try:
            response = requests.get(zip_url, timeout=60)
            response.raise_for_status()  # dispara erro se o link não existir

            # Abre o ZIP diretamente da memória
            with zipfile.ZipFile(io.BytesIO(response.content)) as z:
                csv_name = z.namelist()[0]
                with z.open(csv_name) as f:
                    df = pd.read_csv(f, encoding='CP1252', sep=';', usecols=colunas)

                    # Filtra pelos códigos TUSS
                    df = df[df['CD_PROCEDIMENTO'].astype(str).isin(tuss)]

                    dfs.append(df)

            print(f"{uf} {mes:02d}: {len(df):,} linhas carregadas")

        except Exception as e:
            print(f"\n Falha ao processar {uf} {mes:02d}: {e}")

# Empilha tudo em um único DataFrame
if dfs:
    df_det = pd.concat(dfs, ignore_index=True)
    print(f"\n Base detalhada criada com {len(df_det):,} linhas e {df_det.shape[1]} colunas.")
else:
    print("\n Nenhum arquivo foi carregado com sucesso.")

####################################################################################################
# Salvar arquivo no Drive

# Caminho de saída no Google Drive
#caminho_saida = "hospitalar_DET_2024.xlsx"

# Salva em Excel
#df_det.to_excel(caminho_saida, index=False, engine='openpyxl')


####################################################################################################
# Checagem

df_det.head()
```

```python
%%time

##############################################################################
# BAIXANDO DADOS DE PROCEDIMENTOS HOSPITALARES POR UF - BASES CONSOLIDADAS
##############################################################################

####################################################################################################
# Configuração

# Colunas desejadas
colunas = ['ID_EVENTO_ATENCAO_SAUDE', 'ID_PLANO', 'FAIXA_ETARIA', 'SEXO']

####################################################################################################
# Extração

# Lista para armazenar todos os DataFrames
dfs = []

for uf in ufs:
    print(f"\n########## Iniciando UF: {uf} ##########")
    for mes in range(1, 13):
        zip_url = f"https://dadosabertos.ans.gov.br/FTP/PDA/TISS/HOSPITALAR/2024/{uf}/{uf}_2024{mes:02d}_HOSP_CONS.zip"
        print(f"Baixando: {zip_url}")

        try:
            response = requests.get(zip_url, timeout=60)
            response.raise_for_status()

            with zipfile.ZipFile(io.BytesIO(response.content)) as z:
                csv_name = z.namelist()[0]

                # Detecta encoding
                with z.open(csv_name) as f:
                    raw_data = f.read(50000)
                    encoding_detectado = chardet.detect(raw_data)['encoding']

                # Lê o arquivo completo com o encoding correto
                with z.open(csv_name) as f:
                    df = pd.read_csv(
                        f,
                        encoding=encoding_detectado or 'latin1',
                        sep=';',
                        usecols=colunas,
                        low_memory=False
                    )

                    # Filtra os eventos desejados
                    df = df[df['ID_EVENTO_ATENCAO_SAUDE'].isin(ids_eventos)]

                    dfs.append(df)

            print(f"{uf} {mes:02d}: {len(df):,} linhas carregadas")

        except Exception as e:
            print(f"\n Falha ao processar {uf} {mes:02d}: {e}")

# Empilha tudo
if dfs:
    df_cons = pd.concat(dfs, ignore_index=True)
    print(f"\n Base consolidada criada com {len(df_cons):,} linhas e {df_cons.shape[1]} colunas.")
else:
    print("\n Nenhum arquivo foi carregado com sucesso.")

####################################################################################################
# Salvar arquivo no Drive

# Caminho de saída no Google Drive
#caminho_saida = "hospitalar_CONS_2024.xlsx"

# Salva em Excel
#df_cons.to_excel(caminho_saida, index=False, engine='openpyxl')

####################################################################################################
# Checagem

df_cons.head()
```

## Formatação

### Tratando Variáveis Numéricas

```python
################################################
# CONVERSÃO DE TIPO PARA COLUNAS NUMÉRICAS
################################################

cols_num = ['QT_ITEM_EVENTO_INFORMADO', 'VL_ITEM_EVENTO_INFORMADO']

# Substitui vírgula por ponto e remove espaços
for col in cols_num:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .str.strip()
    )

# Converte para numérico, forçando erros para NaN
df[cols_num] = df[cols_num].apply(
    pd.to_numeric, errors='coerce'
)

print(df[cols_num].dtypes)
print(df[cols_num].describe())
```

### Criando Variáveis

```python
##################################################################################
# CRIANDO VARIÁVEL INDICADORA DE OUTLIER PELO MÉTODO DO INTERVALO INTERQUARTIL
##################################################################################


# Cálculo do IQR
Q1 = df['VL_ITEM_EVENTO_INFORMADO'].quantile(0.25)
Q3 = df['VL_ITEM_EVENTO_INFORMADO'].quantile(0.75)
IQR = Q3 - Q1

##########################
# Limites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

##########################
# Criação da variável OUTLIER
df['OUTLIER'] = df['VL_ITEM_EVENTO_INFORMADO'].apply(lambda x: 1 if x < limite_inferior or x > limite_superior else 0)

##########################
# Exibir limites
print(f'\nLimite inferior:', limite_inferior)
print(f'\nLimite superior:', limite_superior)
print('')

##########################
# Visualizar primeiros resultados
df[['VL_ITEM_EVENTO_INFORMADO', 'OUTLIER']].head()
```

```python
###################################################
# CRIANDO VARIÁVEL COM DESCRIÇÃO DO PROCEDIMENTO
###################################################

df["CD_PROCEDIMENTO"] = df["CD_PROCEDIMENTO"].astype(str)

# Dicionário de mapeamento
map_procedimento = {
    '31309127': 'Parto via vaginal',
    '31309135': 'Parto múltiplo',
    '31309054': 'Cesariana',
    '31309208': 'Cesariana com histerectomia'
}

# Criando nova variável
df["PROCEDIMENTO"] = df["CD_PROCEDIMENTO"].map(map_procedimento)


# checagem de informações de Procedimentos
print(f"\n QT {df['PROCEDIMENTO'].value_counts()}")
print(f"\n PCT {df['PROCEDIMENTO'].value_counts(normalize=True)}")
```

```python
############################################################
# CRIANDO A VARIÁVEL MÊS COM A DESCRIÇÃO
############################################################

# Extrai o número do mês
df['MES_NUM'] = df['ANO_MES_EVENTO'].str.split('-').str[1].astype(int)

# Mapeia para siglas
mapa_meses = {
    1: 'JAN', 2: 'FEV', 3: 'MAR', 4: 'ABR',
    5: 'MAI', 6: 'JUN', 7: 'JUL', 8: 'AGO',
    9: 'SET', 10: 'OUT', 11: 'NOV', 12: 'DEZ'
}

df['MES'] = df['MES_NUM'].map(mapa_meses)
```

```python
##############################
# DADOS DE PACOTES
###############################

# Quantidade total de registros para cálculo do percentual
total = len(df)

# Agrupa e ordena explicitamente pela variável
agrupado = (
    df.groupby('IND_PACOTE')
      .agg(
          QT=('IND_PACOTE', 'count'),
          MEDIA_QT_ITEM=('QT_ITEM_EVENTO_INFORMADO', 'mean'),
          MEDIA_VL_ITEM=('VL_ITEM_EVENTO_INFORMADO', 'mean')
      )
      .sort_index()      # garante 0 primeiro, 1 depois
      .reset_index()
)

# Adiciona coluna percentual
agrupado['PCT'] = agrupado['QT'] / total

# Ajusta a tabela para formato solicitado
tabela_final = pd.DataFrame({
    'SITUAÇÃO': ['IND_PACOTE = 0', 'IND_PACOTE = 1'],
    'QT_REGISTROS': agrupado['QT'],
    'PCT_REGISTROS': agrupado['PCT'],
    'MEDIA_QT_ITEM_EVENTO': agrupado['MEDIA_QT_ITEM'],
    'MEDIA_VL_ITEM_EVENTO': agrupado['MEDIA_VL_ITEM']
})


print(f'\n QT total de registros {total}')
tabela_final
```

```python
######################################
# ANALIDANDO VL_ITEM_EVENTO_INFORMADO
######################################

# Lista de métricas e percentis
metricas = {
    'P90': 0.90,
    'P95': 0.95,
    'P97': 0.97,
    'P99': 0.99,
    'P99_5': 0.995,
    'P99_9': 0.999
}

resultados = []

col = 'VL_ITEM_EVENTO_INFORMADO'
total_registros = len(df)

for nome_metrica, perc in metricas.items():
    # valor do percentil
    valor_percentil = df[col].quantile(perc)

    # quantidade de registros acima
    qt_registros = (df[col] > valor_percentil).sum()

    # percentual em relação ao total
    pct_registros = qt_registros / total_registros

    # média dos valores acima
    media_acima = df.loc[df[col] > valor_percentil, col].mean()

    resultados.append({
        'MÉTRICA': nome_metrica,
        'VL_MEDIO': media_acima,
        'QT_REGISTROS': qt_registros,
        'PCT_REGISTROS': pct_registros
    })

# cria a tabela final
tabela_metricas = pd.DataFrame(resultados)

tabela_metricas

```

## Partos Cesáreos por UF

```python
############################################################
# CÁLCULO DO INDICADOR DE PARTOS CESÁREAS POR UF
############################################################


# SOMANDO A QT DE PARTOS POR TIPO E UF
df_agg = (
    df.groupby(['PROCEDIMENTO', 'UF_PRESTADOR'])['QT_ITEM_EVENTO_INFORMADO']
      .sum()
      .reset_index()
)

########################################
# ORGANIZANDO A BASE

df_pivot = (
    df_agg.pivot(index = 'UF_PRESTADOR',
                 columns = 'PROCEDIMENTO',
                 values = 'QT_ITEM_EVENTO_INFORMADO')
    .reset_index()
)

# Remove o nome do índice
df_pivot.columns.name = None

# Renomeia colunas
df_pivot = df_pivot.rename(columns = {
    'UF_PRESTADOR': 'UF',
    'Cesariana': 'QT_Cesariana',
    'Parto múltiplo': 'QT_Parto_multiplo',
    'Parto via vaginal': 'QT_Parto_via_vaginal'
})

# Preenche possíveis faltas
df_pivot[['QT_Cesariana', 'QT_Parto_multiplo', 'QT_Parto_via_vaginal']] = (
    df_pivot[['QT_Cesariana', 'QT_Parto_multiplo', 'QT_Parto_via_vaginal']].fillna(0)
)


########################################
# CALCULANDO TOTAL DE PARTOS
df_pivot['QT_TOTAL_PARTOS'] = (
    df_pivot['QT_Cesariana']
    + df_pivot['QT_Parto_multiplo']
    + df_pivot['QT_Parto_via_vaginal']
)

########################################
# CALCULANDO INDICADOR TAXA_CESÁREAS

df_pivot['TAXA_CESAREA'] = (df_pivot['QT_Cesariana'] / df_pivot['QT_TOTAL_PARTOS']) * 100


########################################
# Caminho de saída no Google Drive
# file = "taxa_cesareas_uf.xlsx"
# df_pivot.to_excel(file, index=False, engine='openpyxl')

########################################

df_pivot
```

```python
############################################################
# CÁLCULO DO INDICADOR DE PARTOS CESÁREAS POR FAIXA_ETARIA
############################################################


# SOMANDO A QT DE PARTOS POR TIPO E MES
df_agg = (
    df.groupby(['PROCEDIMENTO', 'FAIXA_ETARIA'], observed=False)['QT_ITEM_EVENTO_INFORMADO']
      .sum()
      .reset_index()
)


########################################
# ORGANIZANDO A BASE

df_pivot = (
    df_agg.pivot(index = 'FAIXA_ETARIA',
                 columns = 'PROCEDIMENTO',
                 values = 'QT_ITEM_EVENTO_INFORMADO')
    .reset_index()
)

# Remove o nome do índice
df_pivot.columns.name = None

# Renomeia colunas
df_pivot = df_pivot.rename(columns = {
    'Cesariana': 'QT_Cesariana',
    'Parto múltiplo': 'QT_Parto_multiplo',
    'Parto via vaginal': 'QT_Parto_via_vaginal'
})

# Preenche possíveis faltas
df_pivot[['QT_Cesariana', 'QT_Parto_multiplo', 'QT_Parto_via_vaginal']] = (
    df_pivot[['QT_Cesariana', 'QT_Parto_multiplo', 'QT_Parto_via_vaginal']].fillna(0)
)

########################################
# CALCULANDO TOTAL DE PARTOS
df_pivot['QT_TOTAL_PARTOS'] = (
    df_pivot['QT_Cesariana']
    + df_pivot['QT_Parto_multiplo']
    + df_pivot['QT_Parto_via_vaginal']
)

########################################
# CALCULANDO INDICADOR TAXA_CESÁREAS

df_pivot['TAXA_CESAREA'] = (df_pivot['QT_Cesariana'] / df_pivot['QT_TOTAL_PARTOS']) * 100

########################################
# Caminho de saída no Google Drive
# file = "taxa_cesareas_faixa.xlsx"
# df_pivot.to_excel(file, index=False, engine='openpyxl')

########################################

df_pivot
```

## Despesa por UF

```python
#########################################
# DESPESA POR UF: VALORES MÉDIOS E SOMA
#########################################

# Tabela completa
tabela_completa = (
    df
    .groupby('UF_PRESTADOR')['VL_ITEM_EVENTO_INFORMADO']
    .agg(SOMA='sum', MEDIA='mean')
)

# Tabela sem outliers
tabela_sem_outlier = (
    df[df['OUTLIER'] == 0]
    .groupby('UF_PRESTADOR')['VL_ITEM_EVENTO_INFORMADO']
    .agg(SOMA_SEM_OUTLIER='sum', MEDIA_SEM_OUTLIER='mean')
)

# Junta tudo
tabela_final = (
    tabela_completa
    .join(tabela_sem_outlier, how='left')
    .reset_index()
)

tabela_final['SOMA_MIL'] = tabela_final['SOMA'] / 1000
tabela_final['SOMA_SEM_OUTLIER_MIL'] = tabela_final['SOMA_SEM_OUTLIER'] / 1000

########################################
# Caminho de saída no Google Drive
# file = "despesa_uf.xlsx"
# tabela_final.to_excel(file, index=False, engine='openpyxl')
########################################

tabela_final
```
